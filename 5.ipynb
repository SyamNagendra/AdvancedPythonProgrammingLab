ACCESSING THE ELEMENTS OF A DATAFRAME

```python
import pandas as pd
W=pd.DataFrame({'Name':['john','emily','ryan'],
                 'Age':[25,30,35],
                 'City':['nyc','london','paris']})
x=W.Name      #column access
print(x)
```

```python
y=W['Age']
y
```

```python
row0=W.loc[0]   #row access
row0
```

```python
row1=W.iloc[1]
row1
```

```python
ele0=W.loc[0,'Name']    #element access
ele0
```

```python
ele1=W.iloc[2,2]
ele1
```
