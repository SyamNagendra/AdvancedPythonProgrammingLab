ADDING,DELETING,MODIFYING THE ROWS/COLUMNS IN A DATAFRAME

```python
import pandas as pd
d={'one':[1,2,3],'tow':[20,25,30]}
x=pd.DataFrame(d)
x
```

```python
x['three']=pd.Series([1,2,3])    #column addition
x
```

```python
del x['three']           #column deletion
x
```

```python
x.pop('tow')
x
```

```python
d2={'one':[4,5,6],'tow':[35,40,45]}  #adding rows
y=pd.DataFrame(d2)
x=x.append(y)
x
```

```python
x.drop(2)   #deletion
```

```python
x.drop([0,2])
```
