iterations on df

```python
import pandas as pd
df=pd.DataFrame({'Name':['john','emily','ryan'],
                 'Age':[25,30,35],
                 'City':['nyc','london','paris']})
for index,rows in df.iterrows():
  print(index)
```

```python
for index,rows in df.iterrows():
  print(rows)
```

```python
for row in df.itertuples():
  print(row)
```

```python
for row in df.itertuples():
  print(row.Name)
```

```python
for col,ser in df.items():
  print(col)
```

```python
for col,ser in df.items():
  print(ser)
```

BAR PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/csv/Toyota.csv',index_col=0)
df.dropna(axis=0,inplace=True)
df.head()
```

```python
c=pd.crosstab(index=df['FuelType'],columns='count')
count=dict(c['count'])
fueltype=count.keys()
counts=count.values()
index=np.arange(len(fueltype))
```

```python
plt.bar(index,counts,color=['red','blue','green'])
plt.title('bar plot of fueltypes')
plt.xlabel('fuel types')
plt.ylabel('frequency')
plt.show()
```
