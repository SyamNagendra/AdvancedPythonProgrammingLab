```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/sample_data/Toyota.csv',index_col=0)
df.dropna(axis=0,inplace=True)
df.head()
```

```python
plt.scatter(df['Age'],df['Price'],color='r')
plt.title("Scattern plot of price vs age")
plt.xlabel("Age[months]")
plt.ylabel("price")
plt.show
```
