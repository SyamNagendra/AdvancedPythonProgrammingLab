WAYS TO DEAL WITH NA's IN DATAFRAMES

```python
import pandas as pd
x=pd.DataFrame({'Name':['john','bob','ryan','kim'],
                'Age':[20,25,None,30],
                'City':['nyc','london',None,'seoul']})
x
```

```python
x_droprows=x.dropna(axis=0)     #dropna
display(x_droprows)
```

```python
x_fill=x.fillna('delete')    #fillna
x_fill
```

```python
x['Age']=x['Age'].fillna(x['Age'].mean())  #based on conditon
x
```

GROUPBY OPERATIONS ON DATAFRAME

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
'Kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
display (df)
```

```python
g=df.groupby('Team').groups
g
```

```python
g=df.groupby('Team')
g.get_group('Devils')
```

```python
import numpy as np
g.agg(np.size)
```
