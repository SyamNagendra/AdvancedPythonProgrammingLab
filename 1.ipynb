MERGING DATAFRAMES

```python
import pandas as pd
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                   'Key':['K0','K1','K2','K3']})
display(df1)
df2=pd.DataFrame({'C':['C0','C1','C2','C3'],
                  'B':['C0','C1','C2','C3'],
                   'Key':['K0','K1','K2','K3']})
display(df2)
merge_df=pd.merge(df1,df2,on='Key')
display(merge_df)
```

```python
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Toyota.csv',index_col=0,na_values=('??','????'))
df.dropna(axis=0,inplace=True)
df.head()
```
