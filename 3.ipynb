```python
import numpy as np
a=np.arange(1,10).reshape(3,3)
np.sum(a,axis=None)   #overall sum
```

```python
np.sum(a,axis=0)  #col sum
```

```python
np.sum(a,axis=1)  #row sum
```

```python
b=np.arange(11,20).reshape(3,3)
np.add(a,b)
```

```python
np.multiply(a,b)
np.subtract(a,b)
np.divide(a,b)
np.remainder(a,b)
```

APPLY FUNCTIONS ON DATAFRAMES

```python
import pandas as pd
df=pd.DataFrame({'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]})
res=df.apply(sum)
print(res)
```

```python
#user defined function
def multiply_by_two(column):
  return column*2
ans=df.apply(multiply_by_two)
print(ans )
```

HISTOGRAM

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/csv/Toyota.csv',index_col=0)
df.dropna(axis=0,inplace=True)
df.head()
```

```python
plt.hist(df['KM'],color='green',edgecolor='white',bins=5)
plt.title("histogram of kilometer")
plt.xlabel('kilometer')
plt.ylabel('frequency')
plt.show()
```
